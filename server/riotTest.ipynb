```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()
```

```python
print(os.getenv("test"))
riotAPI = os.getenv("riot_api_key")
```

```python
gameName = 'V1enna'
tagLine = 9849
```

```python
def get_puuid(gameName, tagLine): # 
    link = f'https://asia.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{gameName}/{tagLine}?api_key={riotAPI}'
    print(link)
    response = requests.get(link)
    return response.json()['puuid']
```

```python
user_puuid = get_puuid(gameName, tagLine)
```
